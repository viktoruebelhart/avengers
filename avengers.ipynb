```python
import matplotlib.pyplot as plt

gender_counts = df['Gender'].value_counts()

# Create a figure and axes
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Bar plot
gender_counts.plot(kind='bar', ax=ax[0])
ax[0].set_title('Distribution of Genders (Bar Plot)')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=0)

# Pie chart
ax[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
ax[1].set_title('Distribution of Genders (Pie Chart)')
ax[1].axis('equal')


# Show the plot
plt.tight_layout() # Adjusts subplot params for a tight layout
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the total number of appearances (including duplicates) per year
df_years = df.groupby('Year')['Name/Alias'].count()  # Count of total appearances by year
df_years = df_years.sort_index()  # Sort by year

# Plotting the total number of appearances (or characters) by year
plt.figure(figsize=(10, 6))
df_years.plot(kind='line', marker='o', color='b')
plt.title('Total Number of Avengers Appearances Each Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Appearances', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

# Marvel history begins on August 31, 1939
then there is incorrect Years data

```python
# Count the number of years below 1939
years_below_1939 = df[df['Year'] < 1939]['Year'].count()

print(f"Number of years below 1939: {years_below_1939}")

# Display those years
years_below_1939_values = df[df['Year'] < 1939]['Year'].unique()
print(f"\nYears below 1939: {years_below_1939_values}")
```

```python
# Filter out rows where 'Year' is below 1939
df_1939 = df[df['Year'] >= 1939]

# Verify the change (optional)
years_below_1939 = df[df['Year'] < 1939]['Year'].count()
print(f"Number of years below 1939 after filtering: {years_below_1939}")
```

```python
import matplotlib.pyplot as plt

# Count the total number of appearances (including duplicates) per year
df_years = df_1939.groupby('Year')['Name/Alias'].count()  # Count of total appearances by year
df_years = df_years.sort_index()  # Sort by year

# Plotting the total number of appearances (or characters) by year
plt.figure(figsize=(10, 6))
df_years.plot(kind='line', marker='o', color='b')
plt.title('Total Number of Avengers Appearances Each Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Appearances', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of appearances by year
df_years = df_1939.groupby('Year')['Appearances'].sum()  # Sum of appearances by year
df_years = df_years.sort_index()  # Sort by year

# Plotting the number of members appearances over the years
plt.figure(figsize=(10, 6))
df_years.plot(kind='line', marker='o', color='b')
plt.title('Number of Avengers Members Appearances Each Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Appearances', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
